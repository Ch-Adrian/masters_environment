```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_memory_consumption(mode, source):
    """
    Plot memory consumption for all algorithms for a given mode and source.
    """
    file_data = data[mode][source]
    file_names = file_data["files"]
    algo_names = file_data["names"]
    data_len = file_data["data_len"]

    plt.figure(figsize=(10, 6))
    for i, file in enumerate(file_names):
        df = pd.read_csv(path_to_file + file)
        # Use only up to data_len rows if needed
        df = df.iloc[:data_len]
        plt.plot(df["timestamp"], df["usage_MB"], label=algo_names[i], linewidth=2)
    plt.xlabel("Timestamp")
    plt.ylabel("Memory Usage (MB)")
    plt.title(f"Memory Consumption for {source} (mode {mode})")
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"../plots/summary/memoryPlot_{mode}_{source}.png")
    plt.close()


def plotAll():
    for mode in data.keys():
        for source in data[mode].keys():
                plot_memory_consumption(mode, source)

if __name__ == "__main__":
    # plot_measurements(data["2"]["fb"], 2, 'fb')
    plotAll()
```
